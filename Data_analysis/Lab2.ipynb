```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Frequency = np.array([5.199*10**(14), 5.495*10**(14), 6.881*10**(14), 7.407*10**(14), 8.219*10**(14)])
StopPotential = np.array([0.716, 0.823, 1.347, 1.661, 1.931])
plt.plot(Frequency, StopPotential, marker='o', linestyle='-', color='r', label='Data Points')
plt.title('StopPotential vs. Frequency')
plt.xlabel('Frequency(HZ)')
plt.ylabel('StopPotential(V)')
plt.legend()
for xt, yt in zip(Frequency,StopPotential):
    plt.text(xt, yt, f'({xt:.4}, {yt:.4})', ha='center', va='bottom')
# plt.savefig(r'D:\Desktop\23秋\B39SE\LAB1')
plt.show()
```

```python
slope, intercept = np.polyfit(Frequency, StopPotential, 1)
print("slope:", slope)
```

```python
data = {
    "U": [10.0, 15.0, 20.0, 25.0, 30.0],
    "436nm filter": [20.47, 25.1, 28.63, 31.47, 32.53],
    "546nm filter": [3.47, 3.99, 4.31, 4.65, 4.97]
}
# 获取U的数据和两个滤波器的平均I数据
U = data["U"]
average_436nm = data["436nm filter"]
average_546nm = data["546nm filter"]

# 绘制图形
plt.plot(U, average_436nm, label="436nm filter")
plt.plot(U, average_546nm, label="546nm filter")
plt.xlabel("U (V)")
plt.ylabel("AverageCurrent I (nA)")
plt.title("AverageCurrent I(nA) vs U")
plt.legend()
for xt, yt in zip(U,average_436nm):
    plt.text(xt, yt, f'({xt:.4}, {yt:.4})', ha='center', va='bottom')
for xt, yt in zip(U,average_546nm):
    plt.text(xt, yt, f'({xt:.4}, {yt:.4})', ha='center', va='bottom')
# 显示图形
plt.show()
```

```python
data_0 = {
    "Aperture Diameter (mm)": [2.0, 4.0, 8.0],
    "436nm average current(nA)": [0.6, 2.533, 10.467],
    "546nm average current(nA)": [0.11, 0.362, 1.6]
}

# 获取光圈直径数据和四种光电流数据
aperture_diameter = data_0["Aperture Diameter (mm)"]
average_current_436nm = data_0["436nm average current(nA)"]
average_current_546nm = data_0["546nm average current(nA)"]

# 绘制图形
plt.plot(aperture_diameter, average_current_436nm, label="436nm average current")
plt.plot(aperture_diameter, average_current_546nm, label="546nm average current")
plt.xlabel("Aperture Diameter (mm)")
plt.ylabel("Photocurrent (nA)")
plt.title("Photocurrent with Varying Aperture Diameter")
plt.legend()
for xt, yt in zip(aperture_diameter,average_current_436nm):
    plt.text(xt, yt, f'({xt:.4}, {yt:.4})', ha='center', va='bottom')
for xt, yt in zip(aperture_diameter,average_current_546nm):
    plt.text(xt, yt, f'({xt:.4}, {yt:.4})', ha='center', va='bottom')
# 显示图形
plt.show()
```

```python
data_1 = {
    "Distance (mm)": [400.0, 380.0, 360.0, 340.0, 320.0],
    "Average 436nm filter I(nA)": [9.35, 10.94, 12.4, 15.0, 17.42],
    "Average 546nm filter I(nA)": [1.63, 1.89, 2.15, 2.52, 2.89],
}

# 获取距离数据和四种光电流数据
distance = data_1["Distance (mm)"]
average_current_436nm = data_1["Average 436nm filter I(nA)"]
average_current_546nm = data_1["Average 546nm filter I(nA)"]

# 绘制图形
plt.plot(distance, average_current_436nm, label="436nm filter average current")
plt.plot(distance, average_current_546nm, label="546nm filter average current")
plt.xlabel("Distance (mm)")
plt.ylabel("Photocurrent (nA)")
plt.title("Photocurrent with Varying Distance")
plt.legend()

for xt, yt in zip(distance,average_current_436nm):
    plt.text(xt, yt, f'({xt:.4}, {yt:.4})', ha='center', va='bottom')
for xt, yt in zip(distance,average_current_546nm):
    plt.text(xt, yt, f'({xt:.4}, {yt:.4})', ha='center', va='bottom')

# 显示图形
plt.show()
```
